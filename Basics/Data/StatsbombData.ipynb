# Statsbomb data

Statsbom supplies us with loads of free Data and even provides a easy way to access it via their own package for python and R! <br>
https://github.com/statsbomb/statsbombpy <br>

```python
from statsbombpy import sb
```

```python
# filter for free data 
comps = sb.competitions()
comps = comps[~comps['match_available_360'].isnull()]
comps
```

```python
# All matches in a competition
sb.matches(competition_id=43, season_id=106)
```

```python
# event data for a specific match 
events = sb.events(match_id=3857261)

events
```

```python
# one event type
sb.events(match_id=3857261, split=True, flatten_attrs=False)["shots"]
```
