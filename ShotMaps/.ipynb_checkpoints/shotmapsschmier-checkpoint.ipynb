```python
from PIL import Image

markers={'SavedShot': "triangle-up", 'MissedShots': 'circle', 'BlockedShot': "triangle-down", 'Goal': 'star',
                                'OwnGoal': 'X', 'ShotOnPost': "hexagon"}

# symbols / markers in correct order based on fixed dictionary
symbols = []
for result in rescaled_data.result.unique():
    symbols.append(markers[result])
    
# define colors for both teams
colors = ['red', 'black']

# create scatter 
fig = px.scatter(rescaled_data, x='x', y='y', color='team', symbol='result', size='xG',color_discrete_sequence = colors,
                custom_data=['player', 'xG', 'result'], hover_name= 'player', hover_data = ['xG', 'result'],
                symbol_sequence= symbols, labels=' ')

# adapt the hover to show player and xG value rounded to 2 decimals
fig.update_traces(hovertemplate = "%{customdata[0]}<br>%{customdata[1]}<br>%{customdata[2]}", hoverinfo=None)


# load pitch as image
img = Image.open('mplpitch1.png')

#background image add
# x and y should be substracting the border of the pitch included in the image so that 0,0 is at the corner of the actual pitch
# this then requires an adaption of the size of the image by usually twice what was entered at x and y 
fig.add_layout_image(
    dict(source=img,
         xref='x',
         yref='y',
         x=-3.75,
         y=-3.75,
         sizex=127.5,
         sizey=87.5,
         sizing='stretch',
         opacity=1,
         layer='below')
)

fig.update_layout(
    autosize=True,
    width=130 *5,    # width and height as a multiple of pitch dimension
    height=80*5,     # to maintain dimension / Aspect ratio 
    margin=dict(     # margins of paper
        l=0,
        r=0,
        b=10,
        t=30,        # space for header
    ),
    # hide axes
    xaxis=dict(visible=False,autorange=True),
    yaxis=dict(visible=False,autorange='reversed'),
    # title
    title='Shot map', title_font=dict(color='white', size=25),
    title_x=0.5, title_y=0.975,
    plot_bgcolor='#435348',
    paper_bgcolor='#435348',showlegend=False)

# add scatters in the corners = easiest way to scale plot
fig.add_trace(
    go.Scatter(x=[0,120],
               y=[0,80],
               hoverinfo='skip',
               mode='markers',
               marker=dict(size=0, 
                           color='white',
                          opacity=0)))


# load team logos 
himg = Image.open('bay.png')
aimg = Image.open('bay04.png')

# add logos
fig.add_layout_image(
    dict(source=himg,
         xref='x',
         yref='y',
         x=50,
         y=1,
         sizex=15,
         sizey=15,
         xanchor='center',
         sizing='stretch',
         opacity=0.9,
         layer='below')
)

fig.add_layout_image(
    dict(source=aimg,
         xref='x',
         yref='y',
         x=70,
         y=1,
         sizex=15,
         sizey=15,
         xanchor='center',
         sizing='stretch',
         opacity=0.9,
         layer='below')
)


# determine goal count
ng1 = len(rescaled_data[self.result_col][scaled_shot_data[self.result_col] == 'Goal'][
              scaled_shot_data[self.team_column] == home_team])
+ len(scaled_shot_data[self.result_col][scaled_shot_data[self.result_col] == 'OwnGoal'][
          scaled_shot_data[self.team_column] == away_team])
ng2 = len(scaled_shot_data[self.result_col][scaled_shot_data[self.result_col] == 'Goal'][
              scaled_shot_data[self.team_column] == away_team])
+ len(scaled_shot_data[self.result_col][scaled_shot_data[self.result_col] == 'OwnGoal'][
          scaled_shot_data[self.team_column] == home_team])
```
